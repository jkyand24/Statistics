```python
import seaborn as sns
import matplotlib.pyplot as plt

# load titanic dataset

titanic = sns.load_dataset('titanic')

print(titanic)

# create a pointplot 

sns.pointplot(x = 'sex', y = 'survived', hue = 'pclass', data = titanic)

plt.legend(title = 'Pclass', loc = 'best')

plt.title('Title')
```
