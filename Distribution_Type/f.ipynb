```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

dfn = 5
dfd = 10

num_simulations = 1000
f_values = np.random.f(dfn, dfd, num_simulations)

plt.hist(f_values, bins = 20, alpha = 0.5)
plt.xlabel('F value')
plt.ylabel('Frequency')
plt.title('F distribution')
plt.show()
```
