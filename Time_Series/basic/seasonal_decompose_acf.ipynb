```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf

dates = pd.date_range(start = '2023-01-01', periods = 1000, freq = 'D') # dates - DatetimeIndex
np.random.seed(42)
values = np.random.randn(1000)

data = pd.DataFrame({'Date': dates, 'Value': values})
data.to_csv('./data/sample_time_data.csv', index = False)
data = pd.read_csv('./data/sample_time_data.csv', parse_dates = ['Date'], index_col = 'Date') # 1000 rows  1 columns

plt.plot(data.index, data['Value'])
plt.xlabel('Date\ndata.index')
plt.ylabel("Value\ndata['Value']")
plt.title('Time Series Data')
plt.show()

result = seasonal_decompose(data['Value'], model = 'additive', period = 365) # result - DecomposeResult
result.plot()
plt.show()
```
