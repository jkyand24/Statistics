```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dates = pd.date_range(start = '2023-01-01', periods = 1000, freq = 'D')
np.random.seed(42)
values = np.random.randn(1000)

data = pd.DataFrame({'Date': dates, 'Value': values})

data.loc[100:200, 'Value'] = np.nan
print(data.loc[100:200])
print('\n')

data['Value_filled'] = data['Value'].interpolate(method = 'nearest')
print(data.loc[100:200])
print('\n')

plt.plot(data['Date'], data['Value'], 'o-', label = 'Original')
plt.plot(data['Date'], data['Value_filled'], 'x-', label = 'NN interpolation')

plt.xlabel('Date')
plt.ylabel('Value')
plt.title('NN interpolation')
plt.legend()
plt.show()
```
