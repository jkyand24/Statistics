```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import boxcox

data = pd.DataFrame({'Value': [10, 20, 30, 40, 50]})

transformed_data, lambda_ = boxcox(data['Value'])

plt.plot(data['Value'], label = 'Original')
plt.plot(transformed_data, label = 'Boxcox transformation')
plt.legend()
plt.show()
```

```python
hello = np.random.normal(loc = 3, scale = 1, size = 1000)
hello_sort = np.sort(hello)

plt.plot(hello)
plt.plot(hello_sort)
plt.xlabel('index')
plt.ylabel('value')
plt.show()

plt.hist(hello, bins = 20)
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

```python
hello = hello_sort

hello_log = np.log(hello)

plt.plot(hello)
plt.plot(hello_log)
plt.show()

# log index, value, frequency  value  
plt.hist(hello_log, bins = 20) 
plt.xlabel('np.log(value)')
plt.show()
#  histogram   (left-skewed)
```

```python
hello_exp = np.exp(hello)

plt.plot(hello)
plt.plot(hello_exp)
plt.show()

# exp index, value, frequency  value  
plt.hist(hello_exp, bins = 20) 
plt.xlabel('np.exp(value)')
plt.show()
#  histogram   (right-skewed)
```
