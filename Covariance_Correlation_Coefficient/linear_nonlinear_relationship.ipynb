```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
y = 2 * x + 1
yn = x ** 2

plt.scatter(x, y)
plt.title('Linear Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

plt.scatter(x, yn)
plt.title('Nonlinear Relationship')
plt.xlabel('X')
plt.ylabel('YN')
plt.show()
```
